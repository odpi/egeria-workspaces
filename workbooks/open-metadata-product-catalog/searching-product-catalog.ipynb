# Searching the Open Metadata Digital Product Catalog

This notebook shows a number of options for searching and browsing the **Open Metadata Digital Product Catalog**.  This catalog is build dynamically in the open metadata ecosystem based on the metadata that is loaded and maintained there.  The product catalog includes a folder hierarchy used to organized the products, a glossary describing the concepts used in the catalog and a data dictionary, describing the data values supported by the data products.

We will begin by examining the structure of the digital product catalog and then go on to locate products through various navigational and search operations.  But first we must initialize the python environment.

The code below creates the client used to work with the calogy structure and the products themselves.  This is the Collection Manager Client that calls the [Collection Manager API](https://egeria-project.org/services/omvs/collection-manager/overview/).

## Browsing the catalog

The product catalog is a specialized collection.  The command below retrieves the list of collections with the name *Open Metadata Digital Product Catalog*.

----

```python

results=collectionManagerClient.get_collections_by_name(filter_string="Open Metadata Digital Product Catalog", output_format="MERMAID", report_spec="Collection")
```
